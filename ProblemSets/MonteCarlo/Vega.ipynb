# Problem

## You measure the brightness of a star and find that it is 10,000 +/- 100 times fainter than Vega (magnitude=0). Calculate the starâ€™s magnitude and the error on magnitude using Monte Carlo.

```python
ntrial = 1000000
errors = np.random.normal(loc=0., scale=100., size=ntrial)
rel_flux = 10000. + errors
```

```python
magnitude = -2.5*np.log10(rel_flux)
```

## Plot

```python
plt.figure(figsize=(9,9))
plt.clf()
ax = plt.gca()
sns.histplot(x=magnitude, ax=ax)
#
plotting.set_fontsize(ax, 17)
```

## Stats

```python
avg_mag = np.mean(magnitude)
std_mag = np.std(magnitude)
#
print(f'Average magnitude={avg_mag};  RMS={std_mag}')
```
