```python
#Make a universe where every galaxy is L=10
import numpy as np
import matplotlib.pyplot as plt

#Randomly place galaxies on a Cartesian grid
N=10**5
x=np.random.random_sample(N)
y=np.random.random_sample(N)
z=np.random.random_sample(N)

#Figure out which ones are inside of a sphere
cart_r=np.sqrt(x**2+y**2+z**2)
sphere_r=cart_r[np.where(cart_r < 1)]

#Assign luminosity/flux
Luminosity=10
Flux=Luminosity/sphere_r**2

#Plot
bins = 10**(np.arange(0,20)/4.)
plt.xscale('log')
plt.yscale('log')
plt.hist(Flux,bins=bins)

x=np.arange(100000)+10
y=x**(-1.5)*10**6.2
plt.plot(x,y)
```

```python
#Make a universe where galaxies have luminosities following a powerlaw
import numpy as np
import matplotlib.pyplot as plt

#Randomly place galaxies on a Cartesian grid
N=10**5
x=np.random.random_sample(N)
y=np.random.random_sample(N)
z=np.random.random_sample(N)

#Figure out which ones are inside of a sphere
cart_r=np.sqrt(x**2+y**2+z**2)
sphere_r=cart_r[np.where(cart_r < 1)]

#Draw luminosities
N_galaxies=sphere_r.size
Luminosity=np.random.random_sample(N_galaxies)**(-1/2)

#Assign luminosity
Flux=Luminosity/sphere_r**2

#Plot
bins = 10**(np.arange(0,20)/4.)
plt.xscale('log')
plt.yscale('log')
plt.hist(Flux,bins=bins)

x=np.arange(100000)+5
y=x**(-1.5)*10**5.1
plt.plot(x,y)
```
